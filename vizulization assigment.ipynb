```python
# 5 . Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population
# np.random.seed(25)
#data = {
 #   'Country': ['USA', 'Canada', 'UK',
#'Germany', 'France'],
 #   'Population':
#np.random.randint(100, 1000, 5),
 #   'GDP': np.random.randint(500, 2000,
#5)
#}
#df = pd.DataFrame(data)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate the dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create a bubble chart
plt.figure(figsize=(10, 6))

# Scatter plot with bubble size proportional to population
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population']*10,  # Bubble size proportional to Population
            alpha=0.7, 
            c=df['Population'],      # Color bubbles by population size
            cmap='viridis',          # Use a colormap
            edgecolors='w',          # White edge for each bubble
            linewidth=0.5)

# Annotate each point with the country name
for i in range(len(df)):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], 
             fontsize=12, ha='right', color='black')

# Add labels and title
plt.title('Bubble Chart of Population vs GDP by Country')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.colorbar(label='Population')  # Color bar to show the population scale

plt.show()
```

```python
#1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import numpy as np
from bokeh.plotting import figure, show

# Set x-values from 0 to 10 with a step of 0.1
x = np.arange(0, 10, 0.1)

# Compute y-values as the sine of x
y = np.sin(x * np.pi)

# Create a new figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with circle markers
p.line(x, y, legend_label="y = sin(x)", line_width=2)

# Show the results
show(p)
```

```python
#2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
 #markers based on the 'sizes' and 'colors' columns.
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
sizes = np.random.randint(5, 20, size=100)
colors = np.random.choice(['red', 'blue', 'green', 'yellow'], size=100)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(data=df)

# Create a figure
p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')

# Create a scatter plot
p.scatter('x', 'y', size='sizes', color='colors', source=source)

# Show the plot
show(p)



```

```python
#3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
 # fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)
p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
#4  Create a Bokeh histogram to visualize the distribution of the given data.
#data_hist = np.random.randn(1000)
#hist, edges = np.histogram(data_hist, bins=30)
import numpy as np
from bokeh.plotting import figure, show

# Generate random data
data_hist = np.random.randn(1000)

# Create a histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(width=600, height=400, title="Histogram of Random Data")

# Plot the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white")

# Set axis labels
p.xaxis.axis_label = "Value"
p.yaxis.axis_label = "Frequency"

# Show the plot
show(p)


```
